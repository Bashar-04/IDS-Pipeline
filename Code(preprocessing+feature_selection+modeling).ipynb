# **Model one preparation**



```python
def map_protocol(port):
    if port in [80, 8080, 443]:
        return 'HTTP/HTTPS (TCP)'
    elif port == 53:
        return 'DNS (UDP)'
    elif port == 22:
        return 'SSH (TCP)'
    elif port in [20, 21]:
        return 'FTP (TCP)'
    elif port == 123:
        return 'NTP (UDP)'
    elif port == 161:
        return 'SNMP (UDP)'
    elif port == 445:
        return 'SMB (TCP)'
    elif port == 3389:
        return 'RDP (TCP)'
    elif port in [137, 138, 139]:
        return 'NetBIOS (UDP/TCP)'
    else:
        return 'Other'
```

```python
model1_df = merged_df.copy()
```

```python
model1_df['Traffic_Type'] = model1_df['Destination Port'].apply(map_protocol)
model1_df.drop('Label', axis=1, inplace=True)
```

```python
# Drop zero variance columns
zero_var_cols = model1_df.columns[model1_df.nunique() <= 1].tolist()
model1_df = model1_df.drop(columns=zero_var_cols)
```

# **Feature selection via XGBoost feature importance**

```python
def preprocess_remove_inf(df, target_col):
    # Separate features and target
    X = df.drop(columns=[target_col]).copy()
    y = df[target_col].copy()

    # Find rows with inf or -inf
    mask_inf = X.isin([np.inf, -np.inf]).any(axis=1)

    # Drop rows with inf/-inf in features and corresponding labels
    X = X.loc[~mask_inf].reset_index(drop=True)
    y = y.loc[~mask_inf].reset_index(drop=True)

    # Now handle NaNs if any
    for col in X.select_dtypes(include=[np.number]).columns:
        median_val = X[col].median()
        X[col].fillna(median_val, inplace=True)

    # Label encode categorical features
    for col in X.select_dtypes(include=['object', 'category']).columns:
        le_feat = LabelEncoder()
        X[col] = le_feat.fit_transform(X[col].astype(str))

    # Encode target labels
    le_target = LabelEncoder()
    y_encoded = le_target.fit_transform(y)

    return X, y_encoded, le_target
```

```python
X1_clean, y1_clean, le1 = preprocess_remove_inf(model1_df, 'Traffic_Type')
```

```python
X2_clean, y2_clean, le2 = preprocess_remove_inf(model2_df, 'Suspicious')
```

```python
X3_clean, y3_clean, le3 = preprocess_remove_inf(model3_df, 'Label')
```

```python
X1_train, X1_test, y1_train, y1_test = train_test_split(
    X1_clean, y1_clean, test_size=0.2, random_state=42, stratify=y1_clean
)

d1train = xgb.DMatrix(X1_train, label=y1_train)

params1 = {
    'objective': 'multi:softprob',
    'num_class': len(le1.classes_),
    'eval_metric': 'mlogloss',
    'max_depth': 6,
    'eta': 0.1,
    'seed': 42
}

model1 = xgb.train(params1, d1train, num_boost_round=100)

# === Feature importance extraction ===
importance = model1.get_score(importance_type='gain')
importance_df = pd.DataFrame({
    'feature': list(importance.keys()),
    'importance': list(importance.values())
}).sort_values(by='importance', ascending=False)

print("Feature importance:")
print(importance_df)

# === Select top N features ===
top_n = 10
top_features1 = importance_df['feature'].head(top_n).tolist()
print(f"Top {top_n} features selected:", top_features1)
```

```python

X2_train, X2_test, y2_train, y2_test = train_test_split(
    X2_clean, y2_clean, test_size=0.2, random_state=42, stratify=y2_clean
)

d2train = xgb.DMatrix(X2_train, label=y2_train)

params2 = {
    'objective': 'binary:logistic',
    'eval_metric': 'logloss',
    'max_depth': 6,
    'eta': 0.1,
    'seed': 42
}

model2 = xgb.train(params2, d2train, num_boost_round=100)

# === Feature importance extraction ===
importance = model2.get_score(importance_type='gain')
importance_df = pd.DataFrame({
    'feature': list(importance.keys()),
    'importance': list(importance.values())
}).sort_values(by='importance', ascending=False)

print("Feature importance:")
print(importance_df)

# === Select top N features ===
top_n = 10
top_features2 = importance_df['feature'].head(top_n).tolist()
print(f"Top {top_n} features selected:", top_features2)
```

```python
X3_train, X3_test, y3_train, y3_test = train_test_split(
    X3_clean, y3_clean, test_size=0.2, random_state=42, stratify=y3_clean
)

d3train = xgb.DMatrix(X3_train, label=y3_train)

params3 = {
    'objective': 'multi:softprob',
    'num_class': len(le3.classes_),
    'eval_metric': 'mlogloss',
    'max_depth': 6,
    'eta': 0.1,
    'seed': 42
}

model3 = xgb.train(params3, d3train, num_boost_round=100)

# === Feature importance extraction ===
importance = model3.get_score(importance_type='gain')
importance_df = pd.DataFrame({
    'feature': list(importance.keys()),
    'importance': list(importance.values())
}).sort_values(by='importance', ascending=False)

print("Feature importance:")
print(importance_df)

# === Select top N features ===
top_n = 10
top_features3 = importance_df['feature'].head(top_n).tolist()
print(f"Top {top_n} features selected:", top_features3)
```

# **Model 1**

```python
X_train1_sel = X1_train[top_features1]
X_test1_sel = X1_test[top_features1]

d1train_sel = xgb.DMatrix(X_train1_sel, label=y1_train)
d1test_sel = xgb.DMatrix(X_test1_sel, label=y1_test)

model1_sel = xgb.train(params1, d1train_sel, num_boost_round=100)

# === Predict and evaluate ===
y1_pred_prob = model1_sel.predict(d1test_sel)
y1_pred = y1_pred_prob.argmax(axis=1)

print(f"Accuracy: {accuracy_score(y1_test, y1_pred):.4f}")
print(f"Precision: {precision_score(y1_test, y1_pred, average='weighted'):.4f}")
print(f"Recall: {recall_score(y1_test, y1_pred, average='weighted'):.4f}")
print(f"F1 Score: {f1_score(y1_test, y1_pred, average='weighted'):.4f}")
print("\nClassification report:")
print(classification_report(y1_test, y1_pred, target_names=le1.classes_))
```

# **Model 2**

```python
X_train2_sel = X2_train[top_features2]
X_test2_sel = X2_test[top_features2]

d2train_sel = xgb.DMatrix(X_train2_sel, label=y2_train)
d2test_sel = xgb.DMatrix(X_test2_sel, label=y2_test)

model2_sel = xgb.train(params2, d2train_sel, num_boost_round=100)

# === Predict and evaluate ===
# Predict probabilities for class 1
y2_pred_prob = model2_sel.predict(d2test_sel)

# Convert probabilities to binary predictions using 0.5 threshold
y2_pred = (y2_pred_prob > 0.5).astype(int)

print(f"Accuracy: {accuracy_score(y2_test, y2_pred):.4f}")
print(f"Precision: {precision_score(y2_test, y2_pred):.4f}")
print(f"Recall: {recall_score(y2_test, y2_pred):.4f}")
print(f"F1 Score: {f1_score(y2_test, y2_pred):.4f}")


label_map = {0: "Normal Traffic", 1: "Attack Traffic"}
target_names = [label_map[i] for i in sorted(label_map.keys())]
print("\nClassification report:")
print(classification_report(y2_test, y2_pred, target_names=target_names))
```

```python
X_train3_sel = X3_train[top_features3]
X_test3_sel = X3_test[top_features3]

d3train_sel = xgb.DMatrix(X_train3_sel, label=y3_train)
d3test_sel = xgb.DMatrix(X_test3_sel, label=y3_test)

model3_sel = xgb.train(params3, d3train_sel, num_boost_round=100)

# === Predict and evaluate ===
y3_pred_prob = model3_sel.predict(d3test_sel)
y3_pred = y3_pred_prob.argmax(axis=1)

print(f"Accuracy: {accuracy_score(y3_test, y3_pred):.4f}")
print(f"Precision: {precision_score(y3_test, y3_pred, average='weighted'):.4f}")
print(f"Recall: {recall_score(y3_test, y3_pred, average='weighted'):.4f}")
print(f"F1 Score: {f1_score(y3_test, y3_pred, average='weighted'):.4f}")
print("\nClassification report:")
print(classification_report(y3_test, y3_pred, target_names=le3.classes_))
```
